```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon
```

```python
gdf = gpd.read_file('shapefiles/Asheville_City_Limits.shp')
```

```python
crs={'init':'epsg:4326'}
```

```python
cdf = pd.read_csv('source_files/camp_locations.csv')
bdf = pd.read_csv('source_files/business_locations.csv')
```

```python
cdl = [x.strip('()').split(', ') for x in cdf['Geo']]
bdl = [x.strip('()').split(', ') for x in bdf['Geo']]
```

```python
cdf['latitude'] = [float(x[0]) for x in cdl]
cdf['longitude'] = [float(x[1]) for x in cdl]
bdf['latitude'] = [float(x[0]) for x in bdl]
bdf['longitude'] = [float(x[1]) for x in bdl]
```

```python
cdf.columns, bdf.columns
```

```python
cdf['camp'] = True
bdf['camp'] = False
```

```python
gdf = pd.concat([cdf[['latitude', 'longitude', 'camp']], bdf[['latitude', 'longitude', 'camp']]])
```

```python
geometry=[Point(xy) for xy in zip(gdf['longitude'], gdf['latitude'])]
```

```python
geodata=gpd.GeoDataFrame(gdf,crs=crs, geometry=geometry)
```

```python
fig, ax = plt.subplots(figsize=(20,20))

asheville_shapefile.plot(ax=ax, color='lightgrey', edgecolor='k',alpha=1,linewidth=1)

geodata[geodata['camp']==True].plot(ax=ax, color='blue', markersize=15, label='Encampment');
geodata[geodata['camp']==False].plot(ax=ax, color='red', markersize=15, label='Restaurant/Bar');

fig.suptitle('Restaurants & Bars alongside original "Encampments"', fontsize=36)
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
ax.set_xlabel('Longitude', fontsize=10)
ax.set_ylabel('Latitude', fontsize='medium')
plt.legend(prop={'size':20})
```

```python
geodata['buffer'] = geodata['geometry'].buffer(0.00137)
geodata['buffer1k'] = geodata['geometry'].buffer(0.00274)
```

```python
geodata_c = geodata[geodata['camp']==True]
geodata_b = geodata[geodata['camp']==False]
```

```python
fig, ax = plt.subplots(figsize=(20,20))

asheville_shapefile.plot(ax=ax, color='lightgrey', edgecolor='k',alpha=1,linewidth=1)

geodata_c['buffer'].plot(ax=ax, color='cornflowerblue', markersize=10);
geodata_c[geodata_c['camp']==True].plot(ax=ax, color='blue', markersize=12, label='Encampment');


fig.suptitle('Asheville Encampments: 500 ft radius', fontsize=40)
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
ax.set_xlabel('Longitude', fontsize=10)
ax.set_ylabel('Latitude', fontsize='medium')
```

```python
fig, ax = plt.subplots(figsize=(20,20))

asheville_shapefile.plot(ax=ax, color='lightgrey', edgecolor='k',alpha=1,linewidth=1)

geodata_c['buffer1k'].plot(ax=ax, color='cornflowerblue', markersize=10);
geodata_c[geodata_c['camp']==True].plot(ax=ax, color='blue', markersize=15, label='Encampment');


fig.suptitle('Asheville Encampments: 1000 ft radius', fontsize=40)
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
ax.set_xlabel('Longitude', fontsize=10)
ax.set_ylabel('Latitude', fontsize='medium')
```
